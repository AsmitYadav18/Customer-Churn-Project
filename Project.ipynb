```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(10,6))
sns.scatterplot(data=dataset,x='Credit_Limit',y='Avg_Utilization_Ratio',hue='Attrition_Flag',palette={'Existing Customer':'blue', 'Attrited Customer':'red'},alpha=0.6)
plt.title('Credit Limit vs Avg Utilization Ratio by Attrition')
plt.xlabel('Credit Limit')
plt.ylabel('Average Utilization Ratio')
plt.show()
```

```python
pivot = dataset.pivot_table(index='Total_Relationship_Count',columns='Months_Inactive_12_mon',values='Attrition_Flag',aggfunc=lambda x: (x=='Attrited Customer').mean() * 100)

plt.figure(figsize=(10,6))
sns.heatmap(pivot, annot=True, fmt=".1f", cmap="Reds")
plt.title('Attrition Rate by Total Relationships & Months Inactive')
plt.xlabel('Months Inactive (Past 12 Months)')
plt.ylabel('Total Relationship Count')
plt.show()
```

```python
bins=[0,25,50,75,100, dataset["Total_Trans_Ct"].max()]
label = ["Very Low","Low","Medium","High","Very High"]
dataset["Trans_Count"]= pd.cut(dataset["Total_Trans_Ct"], bins=bins, labels=label, include_lowest=True)
churn_rate = dataset.groupby("Trans_Count")["Attrition_Flag"].apply(lambda x: (x=="Attrited Customer").mean()*100)
print(churn_rate)
```

```python
churn_rate.plot(kind='bar', color='skyblue', figsize=(8,5))
plt.ylabel('Churn Rate (%)')
plt.title('Churn Rate by Transaction Count Bin')
plt.show()
```
